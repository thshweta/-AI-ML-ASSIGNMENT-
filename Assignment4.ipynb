# Using the numpy

```python
import numpy as np

age_array = np.array(df["Age"])
print(age_array)
```

```python
average_age = np.mean(age_array)
print(f"Average Age: {average_age}")
```

```python
median_age = np.median(age_array)
print(f"Median Age: {median_age}")
```

```python
income_array = np.array(df["Income"])
std_income = np.std(income_array)
print(f"Standard Deviation of Income: {std_income}")
```

```python
credit_scores = np.array(df["Credit Score"])
print(credit_scores)
max_credit = np.max(credit_scores)
print(max_credit)

```

```python
older_than_40 = np.sum(age_array > 40)
percentage = (older_than_40 / len(age_array)) * 100

print(f"Percentage of people older than 40: {percentage:.2f}%")
```

```python
normalized_income = (income_array - np.min(income_array)) / (np.max(income_array) - np.min(income_array))
print(normalized_income)
```

```python
health_index_array = np.array(df["Health Index"])
count_high_health = np.sum(health_index_array > 70)

print(f"People with Health Index > 70: {count_high_health}")
```

```python
ones_array = np.ones_like(age_array)
print(ones_array)
```

```python
correlation = np.corrcoef(age_array, income_array)[0, 1]
print(f"Correlation between Age and Income: {correlation:.2f}")
```

```python
country_array = np.array(df["Country"])
unique_countries = np.unique(country_array)
print(unique_countries)
```

```python
exercise_array = np.array(df["Exercise Frequency"])
variance_exercise = np.var(exercise_array)

print(f"Variance of Exercise Frequency: {variance_exercise}")
```

```python
random_ages = np.random.randint(20, 60, size=10)
print(random_ages)
```

```python
average_income = np.mean(income_array)
income_array[income_array < 20000] = average_income

print(income_array)
```

```python
percentile_90 = np.percentile(credit_scores, 90)
print(f"90th Percentile of Credit Score: {percentile_90}")
```

# USing Matplotlib and seaborn

```python
import matplotlib.pyplot as plt

plt.hist(df["Age"], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Age")
plt.show()
```

```python
country_counts = df["Country"].value_counts()

plt.figure(figsize=(10,5))
plt.bar(country_counts.index, country_counts.values, color='orange')
plt.xlabel("Country")
plt.ylabel("Number of People")
plt.title("People Count by Country")
plt.xticks(rotation=45)
plt.show()
```

```python
gender_counts = df["Gender"].value_counts()

plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['blue', 'pink'])
plt.title("Gender Distribution")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x=df["Income"], color='green')
plt.title("Box Plot of Income")
plt.show()
```

```python
plt.scatter(df["Age"], df["Income"], color='red', alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Age vs. Income Scatter Plot")
plt.show()
```

```python
# Select only numerical features for correlation
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
age_income = df.groupby("Age")["Income"].mean()

plt.plot(age_income.index, age_income.values, marker='o', linestyle='-', color='purple')
plt.xlabel("Age")
plt.ylabel("Average Income")
plt.title("Average Income by Age")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x=df["Gender"], y=df["Credit Score"], palette="coolwarm")
plt.title("Credit Score Distribution by Gender")
plt.show()
```

```python
sns.pairplot(df[["Age", "Income", "Credit Score"]])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df["Marital Status"], palette="viridis")
plt.title("Marital Status Count")
plt.show()
```

```python
sns.kdeplot(df["Health Index"], fill=True, color='blue')
plt.title("Density Plot of Health Index")
plt.show()
```

```python
sns.scatterplot(x=df["Exercise Frequency"], y=df["Health Index"], hue=df["Gender"], alpha=0.7)
plt.title("Exercise Frequency vs. Health Index")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df["Car Ownership"], hue=df["Gender"], palette="Set2")
plt.title("Car Ownership by Gender")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df["Occupation"], y=df["Income"], palette="muted")
plt.xticks(rotation=45)
plt.title("Income Distribution by Occupation")
plt.show()
```

```python
sns.lmplot(x="Age", y="Credit Score", data=df, aspect=1.5, scatter_kws={'alpha':0.5})
plt.title("Age vs Credit Score with Trend Line")
plt.show()
```

```python
edu_income = df.groupby("Education Level")["Income"].mean().sort_values()

plt.figure(figsize=(10,5))
sns.barplot(x=edu_income.index, y=edu_income.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Education Level")
plt.ylabel("Average Income")
plt.title("Average Income by Education Level")
plt.show()
```

```python
plt.hist(df["Number of Children"], bins=6, color='teal', edgecolor='black')
plt.xlabel("Number of Children")
plt.ylabel("Frequency")
plt.title("Distribution of Number of Children")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hexbin(df["Diet Quality Score"], df["Credit Score"], gridsize=25, cmap="Purples", alpha=0.8)
plt.colorbar(label="Count")
plt.xlabel("Diet Quality Score")
plt.ylabel("Credit Score")
plt.title("Hexbin Plot: Diet Quality Score vs. Credit Score")
plt.show()
```

```python
sns.jointplot(x="Exercise Frequency", y="Health Index", data=df, kind="reg", color="green")
plt.show()
```

```python
sns.regplot(x=df["Income"], y=df["Credit Score"], scatter_kws={"alpha":0.5}, line_kws={"color":"red"})
plt.title("Income vs Credit Score Regression Plot")
plt.show()
```
