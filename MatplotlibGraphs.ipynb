```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Generate sample data
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C', 'D', 'E'], 100),  # Repeat Categories
    'Sales': np.random.randint(100, 500, 100),  # Make Sales 100 rows
    'Marks': np.random.randint(40, 100, 100),
    'Revenue': np.random.randint(1000, 10000, 100),
    'Advertising': np.random.randint(500, 5000, 100),
    'Age': np.random.randint(18, 60, 100),
    'Salary': np.random.randint(20000, 100000, 100),
    'Height': np.random.randint(150, 200, 100),
    'Weight': np.random.randint(50, 100, 100)
})

print(data.head())  # Display first 5 rows

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Line Plot

```python
plt.figure(figsize=(6,4))
plt.plot(range(1, 11), np.random.randint(100, 200, 10), marker='o', linestyle='-')
plt.title("Stock Market Trend Over 10 Days")
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.grid()
plt.show()
```

## Bar Chart

```python
plt.figure(figsize=(6,4))
plt.bar(data['Category'], data['Sales'], color='skyblue')
plt.title("Sales Comparison")
plt.xlabel("Product Categories")
plt.ylabel("Sales")
plt.show()
```

## Histogram

```python
plt.figure(figsize=(6,4))
plt.hist(data['Marks'], bins=10, color='orange', edgecolor='black')
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(6,4))
plt.scatter(data['Advertising'], data['Revenue'], color='red', alpha=0.6)
plt.title("Advertising Budget vs Revenue")
plt.xlabel("Advertising Budget ($)")
plt.ylabel("Revenue ($)")
plt.show()
```

## Pie Chart

```python
plt.figure(figsize=(6,4))
plt.pie(data['Sales'], labels=data['Category'], autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple', 'yellow'])
plt.title("Market Share of Different Brands")
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(6,4))
plt.boxplot(data['Salary'])
plt.title("Salary Distribution")
plt.ylabel("Salary ($)")
plt.show()
```

## Stacked Bar Chart

```python
age_groups = ['0-18', '19-35', '36-50', '51+']
values = np.random.randint(100, 500, (4, 3))
plt.figure(figsize=(6,4))
plt.bar(age_groups, values[:,0], label="City A")
plt.bar(age_groups, values[:,1], bottom=values[:,0], label="City B")
plt.bar(age_groups, values[:,2], bottom=values[:,0]+values[:,1], label="City C")
plt.title("Population Distribution by Age")
plt.xlabel("Age Groups")
plt.ylabel("Population Count")
plt.legend()
plt.show()
```

## HeatMap

```python
heatmap_data = np.random.rand(10, 10)
plt.figure(figsize=(6,4))
plt.imshow(heatmap_data, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title("Temperature Variations Across City")
plt.show()
```

## Polar Plot

```python
theta = np.linspace(0, 2*np.pi, 8)
r = np.random.randint(10, 30, 8)
plt.figure(figsize=(6,4))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, marker='o')
ax.set_title("Wind Speed & Direction")
plt.show()
```

## 3D surface plot

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(np.linspace(-5, 5, 30), np.linspace(-5, 5, 30))
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("Terrain Elevation")
plt.show()
```
