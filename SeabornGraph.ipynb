```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Generate sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Sales': np.random.randint(100, 500, 5),
    'Marks': np.random.randint(40, 100, 5),  # Reduce to 5 rows
    'Revenue': np.random.randint(1000, 10000, 5),
    'Advertising': np.random.randint(500, 5000, 5),
    'Age': np.random.randint(18, 60, 5),
    'Salary': np.random.randint(20000, 100000, 5),
    'Height': np.random.randint(150, 200, 5),
    'Weight': np.random.randint(50, 100, 5)
})

print(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Line Plot - Yearly Sales Trend
plt.figure(figsize=(6,4))
sns.lineplot(x=range(1, 11), y=np.random.randint(100, 200, 10))
plt.title("Yearly Sales Trend")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=data['Category'], y=data['Sales'])
plt.title("Average Scores in Different Subjects")
plt.show()
```

```python
#Histogram - Housing Prices Distribution
plt.figure(figsize=(6,4))
sns.histplot(data['Marks'], bins=10, kde=True, color='green')
plt.title("Housing Prices Distribution")
plt.show()
```

```python
# 4. Scatter Plot - Height vs Weight
plt.figure(figsize=(6,4))
sns.scatterplot(x=data['Height'], y=data['Weight'], hue=data['Age'], palette='coolwarm')
plt.title("Height vs Weight")
plt.show()
```

```python
# 5. Box Plot - Exam Scores
plt.figure(figsize=(6,4))
sns.boxplot(y=data['Marks'], color='purple')
plt.title("Exam Scores Distribution")
plt.show()
```

```python
# 6. Violin Plot - Income Distribution
plt.figure(figsize=(6,4))
sns.violinplot(y=data['Salary'], color='red')
plt.title("Income Distribution")
plt.show()
```

```python
# 7. Pair Plot - Multi-variable Analysis
sns.pairplot(data[['Marks', 'Salary', 'Height', 'Weight']])
plt.show()
```

```python
# 8. Heatmap - Correlation Matrix
plt.figure(figsize=(6,4))
sns.heatmap(data[['Marks', 'Salary', 'Height', 'Weight']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Variables")
plt.show()
```

```python

# 9. Swarm Plot - Customer Ratings
plt.figure(figsize=(6,4))
sns.swarmplot(x=data['Category'], y=data['Sales'], color='blue')
plt.title("Customer Ratings for Products")
plt.show()
```

```python
# 10. Joint Plot - Age vs Spending
sns.jointplot(x=data['Age'], y=data['Salary'], kind='hex', cmap='coolwarm')
plt.show()
```
